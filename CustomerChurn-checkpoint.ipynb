```python
def printUniqueValues(df):     #function to check for unique values
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
printUniqueValues(df1)   #display unique values in dataframe 
```

```python
df1=df1.replace('No internet service','No')   #replacing synonymous values
df1=df1.replace('No phone service','No')
```

```python
printUniqueValues(df1) 
```

```python
binValueColumns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']     #label encoding
```

```python
for col in binValueColumns:
         df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:                             
    print(f'{col}: {df1[col].unique()}')     #display all unique values
```

```python
 df1.TotalCharges=pd.to_numeric(df1.TotalCharges)  
```

```python
#Data Cleaning Continues
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df1.gender.unique()
```

```python
df2 = pd.get_dummies(data=df1, dtype=int, columns=['InternetService','Contract','PaymentMethod'])    #one hot encoding
df2.columns
```

```python
df2.sample(7)
```

```python
df2.dtypes
```

```python
scaledCols = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler                   #scaling values in range
scaler = MinMaxScaler()
df2[scaledCols] = scaler.fit_transform(df2[scaledCols])
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

```python
#Training and testing

X = df2.drop('Churn',axis='columns')    #drop feature
y = df2['Churn']

from sklearn.model_selection import train_test_split                                     #necessary libraries
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)    #training and testing specifications
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python

#Building ANN model in tensorflow/keras 
import tensorflow as tf
from tensorflow import keras


model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# opt = keras.optimizers.Adam(learning_rate=0.01)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=50
        )
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
